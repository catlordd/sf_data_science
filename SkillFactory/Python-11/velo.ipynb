```python
import pandas as pd
import numpy as np


# Reading the data
bike_data = pd.read_csv('data/citibike-tripdata.csv', sep=',')

# bike_data.info()

# Exercise 6.1
None_mean = 0 
for x in bike_data['start station id'].notnull():
    if x:
        None_mean += 1 

#display(len(bike_data['start station id'].notnull()) - None_mean)

# Exercise 6.2

bike_data.dtypes

# Exercise 6.3

bike_data['start station id'].mode()

# Exercise 6.4

bike_data['bikeid'].mode()

# Exercise 6.5

Subscriber = len(bike_data[bike_data['usertype'] == "Subscriber"])
Customer = len(bike_data[bike_data['usertype'] == "Customer"])
total_usertypes = len(bike_data['usertype'])
if Customer > Subscriber:
    print("Customer")
    print(round(Customer/total_usertypes, 2))
else:
    print("Subscriber")
    print(round(Subscriber/total_usertypes, 2))
    
# Exercise 6.6

g_man = len(bike_data[bike_data['gender'] == 1])
g_woman = len(bike_data[bike_data['gender'] == 2])

if g_man > g_woman:
    print("Man:" + str(g_man))
else:
    print("Woman" + str(g_woman))
    
# Exercise 6.7



# min_st_series = bike_data['end station name'].sort_values()

# display(bike_data['end station name'].value_counts())

# Exercise 6.8 (Удалить ненужные дублирующие признаки)

bike_data = bike_data.drop(['start station name', 'end station name'], axis=1)
# columns_ = len(bike_data.columns)

# Exercise 6.9 (Число поездок людей с возрастом 60+)

bike_data['age'] = bike_data['birth year'].apply(lambda x: 2018-x)
bike_data = bike_data.drop(columns='birth year')
bike_data[bike_data['age'] > 60].count()

# Exercise 6.10 (Длительность поездки)

bike_data['starttime'] = pd.to_datetime(bike_data['starttime'])
bike_data['stoptime'] = pd.to_datetime(bike_data['stoptime'])

bike_data['trip duration'] = bike_data['stoptime'] - bike_data['starttime']
bike_data['trip duration'].dt.seconds.iloc[3] / 60

# Exercise 6.11 (Сколько поездок началось в выходные)

bike_data['weekend'] = bike_data['starttime'].dt.day_of_week.apply(lambda x : 1 if x>4 else 0)
bike_data['weekend'].value_counts().loc[1]

# Exercise 6.12 (Создайте признак времени суток поездки)

def to_time_of_day(arg):
    if arg>=0 and arg<=6:
        return 'night'
    elif arg>6 and arg<=12:
        return 'morning'  
    elif arg>12 and arg<=18:
        return 'day'
    else:
        return 'evening'

bike_data['time_of_day'] = bike_data['starttime'].dt.hour.apply(to_time_of_day)

# Во сколько раз количество поездок, совершённых днём, больше, 
# чем количество поездок, совёршенных ночью, за представленный в данных период времени?
 
round(bike_data['time_of_day'].value_counts().loc['day']/bike_data['time_of_day'].value_counts().loc['night'])

bike_data['starttime'].dt.day
```
